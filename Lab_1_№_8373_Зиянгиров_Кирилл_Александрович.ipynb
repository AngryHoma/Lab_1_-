```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

shop = pd.read_csv('Customers1.csv')
shop.head()
```

<font color='Blue'>СКО:</font>

```python
sko_Gender = shop['Gender'].std()
print(sko_Gender)
```
